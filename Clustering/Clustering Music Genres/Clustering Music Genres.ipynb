Bu proje, müzik türlerini tahminlemek için kullanılan bir kümeleme modeli geliştirmeyi amaçlamaktadır. Veri seti olarak, Spotify web API'sinden elde edilen müzik parçalarının özelliklerini içeren bir dataset kullanılmıştır.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("Spotify-2000.csv")
df.head()
```

```python
df = df.drop("Index", axis=1)
df.head()
```

```python
numeric = df.select_dtypes(include=['int64', 'float64'])

numeric.head()
```

```python
numeric.corr()
```

### Ses Özelliklerine GöreKümeleme Analizi

```python
df2= df[["Beats Per Minute (BPM)","Loudness (dB)","Liveness","Valence","Acousticness","Speechiness"]]
```

```python
from sklearn.preprocessing import MinMaxScaler

for i in df.columns:
    MinMaxScaler(i)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(df2)
```

```python
df["Music Segments"] = clusters
MinMaxScaler().fit_transform(df["Music Segments"].values.reshape(-1, 1))
df["Music Segments"] = df["Music Segments"].map({1: "Küme01", 2: 
    "Küme02", 3: "Küme03", 4: "Küme04", 5: "Küme05", 
    6: "Küme06", 7: "Küme07", 8: "Küme08", 
    9: "Küme09", 10: "Küme10"})
```

```python
df.head()
```

```python
import plotly.graph_objects as go
PLOT = go.Figure()
for i in list(df["Music Segments"].unique()):
    

    PLOT.add_trace(go.Scatter3d(x = df[df["Music Segments"]== i]['Beats Per Minute (BPM)'],
                                y = df[df["Music Segments"] == i]['Energy'],
                                z = df[df["Music Segments"] == i]['Danceability'],                        
                                mode = 'markers',marker_size = 6, marker_line_width = 1,
                                name = str(i)))
PLOT.update_traces(hovertemplate='Beats Per Minute (BPM): %{x} <br>Energy: %{y} <br>Danceability: %{z}')

    
PLOT.update_layout(width = 800, height = 800, autosize = True, showlegend = True,
                   scene = dict(xaxis=dict(title = 'Beats Per Minute (BPM)', titlefont_color = 'black'),
                                yaxis=dict(title = 'Energy', titlefont_color = 'black'),
                                zaxis=dict(title = 'Danceability', titlefont_color = 'black')),
                   font = dict(family = "Gilroy", color  = 'black', size = 12))
```
