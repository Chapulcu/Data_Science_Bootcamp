```python
import yfinance as yf
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python

# Apple için hisse senedi sembolünü tanımla
ticker = "AAPL"

# Ticker nesnesi oluştur
apple = yf.Ticker(ticker)

# Geçmiş piyasa verilerini al
hist = apple.history(start="2023-01-01", end="2024-10-10")

# 'Date' sütununu normal bir sütun yapmak için indeksi sıfırla
hist = hist.reset_index()

# 'Date' sütununu datetime formatına çevir ve sadece tarih kısmını al
hist['Date'] = pd.to_datetime(hist['Date']).dt.date

# Verileri CSV dosyasına kaydet
hist.to_csv("apple_stock_data.csv", index=False)

# Verinin ilk birkaç satırını göster
hist.head()
```

```python
hist['Date'].unique()
```

```python
figure = go.Figure(data=[go.Candlestick(x=hist.index,
                                        open=hist["Open"], high=hist["High"],
                                        low=hist["Low"], close=hist["Close"])])
figure.update_layout(title = "Apple Hisse Fiyat Analizi", xaxis_rangeslider_visible=False)
figure.show()
```

```python
hist.info()
```

```python
hist.select_dtypes(include=[np.number]).corr()
```

```python
#pip install autots
```

```python
from autots import AutoTS
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')
model = model.fit(hist, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
