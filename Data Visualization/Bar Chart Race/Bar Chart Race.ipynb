```python
import bar_chart_race as bcr
df = bcr.load_dataset('covid19_tutorial')
df.head()
```

```python
#pip install ffmpeg-python
```

```python
from IPython.display import Image

gif = bcr.bar_chart_race(
    df=df,
    filename='bar_chart_race.gif',  # GIF formatı kullanıyoruz
    orientation='h',
    sort='desc',
    n_bars=10,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=10,
    period_length=500,
    interpolate_period=False,
    scale='linear',
    filter_column_colors=False,
    title='Covid19',
    dpi=144
)

Image(filename='bar_chart_race.gif')
```
